```python
import pandas as pd
from sklearn.preprocessing import RobustScaler

data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)
print("Raw Data Frame")
print(df)


# Task1 - Encoding the text to Numberic or Binary

df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

df = pd.get_dummies(df, columns=['City'], drop_first=True) # avoiding the dummy variable trap

size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)

# Task2 - Scaling the numeric data. RobustScaler is selected here, as the Fare column was having outlier and same scaling method applied for age as well


scaler = RobustScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print("\n\nModified Data Frame")
print(df)
```
