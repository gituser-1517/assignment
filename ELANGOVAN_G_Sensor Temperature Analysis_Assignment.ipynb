```python
import numpy as np

data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1],
                 [17.3, 22.8, 30.5, 26.4, 21.9],
                 [33.1, 29.6, 18.4, 24.0, 27.8],
                 [20.2, 23.5, 31.9, 28.1, 22.6]])

# Task 1 Print the shape of data. Then compute and print the mean temperature per station (one value per row).
print("==================Task1==================")
print("The Shape of the matrix is:",data.shape)
print("Mean Temperature per station:",data.mean(axis=1))

# Task 2 Using a boolean mask, extract all temperature readings above 28.0°C and print them as a 1D array.
print("\n\n==================Task2==================")
high_temp = data > 28
print("Boolean result(True or False) for the temperature readings above 28.0°C is below:\n", high_temp.reshape(-1))

# Task 3 Normalize the entire data array to the range [0, 1] using the formula below and print the result rounded to 2 decimal places:
print("\n\n==================Task3==================")
normalized = (data - data.min()) / (data.max() - data.min())
print("Normalized value of the array with the result rounded to 2 decimal places is below:\n",np.round(normalized,2))
```
